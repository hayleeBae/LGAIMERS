```python
# 이상치 확인 및 처리
train_data.boxplot(figsize = (10,4)) # box plot
plt.show()
```

```python
# 전체 평균값 시각화
visual_data = train_data.drop(columns=['ID','제품'])

plt.figure(figsize = (12,6))
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(60)) # 7일 간격으로 눈금 표시
visual_data.iloc[:,5:].mean().plot()
plt.show()
```

```python
# 문제가 될 것 같은 구간1
visual_data.iloc[:, 30:42].mean().plot()
plt.show()
```

```python
# 이상치 보정 전의 구간 그래프
visual_data.iloc[:, 5:50].mean().plot()
plt.show()
```
